# 2. Sales Patterns Over Time

1.Aggregate new_policy_count by year_month to track overall company sales trend.

```python
import matplotlib.pyplot as plt
import seaborn as sns

monthly_sales = df_train.groupby('year_month')['new_policy_count'].sum().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(x='year_month', y='new_policy_count', data=monthly_sales, marker='o')
plt.title("Total New Policies Sold Per Month")
plt.xlabel("Month")
plt.ylabel("Total Policies Sold")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

#### 📈 Monthly Sales Trend (Jan 2023 – Jul 2024) Insight

The line chart below aggregates the total number of new insurance policies sold each month. We observe a **steady and significant increase in monthly sales**, growing from ~13,000 in early 2023 to over 18,000 by mid-2024.

Key takeaways:
- The company has shown **consistent growth in customer acquisition**.
- Short-term dips in March and October 2023 may indicate seasonal effects or operational slowdowns.
- The **notable surge in June–July 2024** could suggest a successful marketing push or peak season behavior.

This trend provides a strong signal of overall company health and sets the context for identifying agents whose performance deviates from the norm.


2. Active Agents Over Time - 
Count how many agents made sales per month

```python
active_agents = df_train[df_train['new_policy_count'] > 0].groupby('year_month')['agent_code'].nunique().reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(x='year_month', y='agent_code', data=active_agents)
plt.title("Number of Active Agents Per Month (Sold ≥1 Policy)")
plt.xlabel("Month")
plt.ylabel("Active Agents")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

3. NILL Agents Over Time - 
Count how many agents had zero sales in each month.

```python
nill_agents = df_train[df_train['new_policy_count'] == 0].groupby('year_month')['agent_code'].nunique().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(x='year_month', y='agent_code', data=nill_agents, marker='o', color='red')
plt.title("Number of NILL Agents Per Month (Sold 0 Policies)")
plt.xlabel("Month")
plt.ylabel("NILL Agents")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

#### 🚫 NILL Agents Trend (Zero Sales Per Month) Insight

The chart below tracks the number of agents who sold **zero policies per month**. Despite the company’s overall sales growth, the number of inactive agents remained **consistently high**, with **peaks exceeding 100 NILL agents** in early 2024.

Key Observations:
- **No clear downward trend** in underperformance.
- **Volatile monthly changes** suggest a lack of consistent agent management strategies.
- High NILL counts alongside sales growth could indicate **dependence on a few high performers**, masking broader productivity issues.

📌 This insight highlights the need for targeted interventions such as training, lead redistribution, or performance-based incentives to address inactive agents.


#  3. Multivariate Analysis

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Specify the selected columns
selected_cols = [
    'unique_proposal',
    'unique_quotations',
    'unique_customers',
    'new_policy_count',
    'number_of_policy_holders',
    'net_income'
]

# Compute correlation matrix for selected columns only
corr = df_train[selected_cols].corr()

# Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix of Selected Features")
plt.tight_layout()
plt.show()

```

#### 🔗 Multivariate Insights: What Drives Net Income? Insight

The correlation heatmap below explores the relationships among key business metrics. Notable takeaways include:

- **Proposal activity** is the strongest positive driver of both new policies and income.
- A surprising finding: having more **policyholders** or even more **new policies** doesn't strongly correlate with **net income**, possibly due to differences in policy value, cost structures, or payment cycles.
- The pipeline appears to flow logically from **quotations → proposals → policies**, but not all stages equally impact revenue.

💡 This suggests optimizing **proposal quality and conversion** could be more effective than merely increasing the policy count.


```python
pair_cols = ['new_policy_count', 'unique_proposal', 'unique_quotations',
             'unique_customers', 'number_of_policy_holders', 'net_income']

sns.pairplot(df_train[pair_cols])
plt.suptitle("Pairwise Feature Relationships", y=1.02)
plt.show()
```

### 📈 Feature Correlations and Distributions

#### 🔗 Strong Linear Relationships
- There's a **very strong positive linear correlation** between `unique_proposal` and `unique_quotations`, indicating that agents who create more proposals also generate quotations at a consistent rate.
- Similarly, `unique_customers` is strongly correlated with both `proposal` and `quotation` metrics, suggesting that **customer acquisition directly drives proposal and quotation activity**.

#### 🧭 Distinct Agent Clusters
- Several scatter plots display **discrete clusters or patterns**, potentially representing different agent segments or behavioral types.
- Diagonal line patterns involving `tenure_months` suggest **cohorts of agents hired at specific times**, reflecting distinct tenure groups.

#### 📉 Non-Linear Relationships
- The relationship between `net_income` and activity metrics (`unique_proposal`, `unique_quotations`, `unique_customers`) exhibits a **non-linear pattern**.
- After certain thresholds, increased activity **doesn't translate into proportionally higher income**, possibly due to diminishing returns or workload saturation.

#### 📊 Multimodal Distributions
- The histogram of `tenure_months` shows **multiple peaks**, hinting at hiring waves or seasonal recruitment trends.
- `number_of_policy_holders` appears **skewed with multiple modes**, indicating varied client portfolio sizes across agents.

#### 📐 Right-Skewed Metrics
- Most performance metrics (`unique_proposal`, `unique_quotations`, `unique_customers`, `net_income`) show **right-skewed distributions**.
- This reflects a **Pareto principle pattern**, where a small subset of agents drive most of the performance.

#### 🧱 Ceiling Effects
- Some features exhibit **ceiling effects**, where beyond a certain point, increasing one metric doesn't yield growth in the other—possibly due to capacity limits or operational constraints.

---

### 🛠️ Implications for Model Building

#### 🧪 Feature Engineering Opportunities
- Strong linear correlations suggest the value of **ratio features**, such as **conversion rates** between quotations, proposals, and policies.
- The **non-linear trends** support exploring polynomial or interaction features.

#### 🧩 Data Segmentation Potential
- The presence of clusters indicates the potential benefit of **segmented modeling** (e.g., clustering-based models or separate models per agent group).

#### ⚠️ Outlier Considerations
- Several extreme values exist across features. These **outliers may need to be capped, scaled, or modeled separately**.

#### 🚨 Multicollinearity Warning
- Strong correlations among features (e.g., proposals and quotations) point to **multicollinearity**, which could degrade model interpretability and performance. Dimensionality reduction or feature selection is recommended.

#### 🎯 Target Variable Insights
- While activity metrics show correlation with `net_income`, they **don't fully explain income variation**, indicating that other unobserved or latent factors influence performance quality.

# 5. Innovative EDA 

Cluster agents based on performance patterns

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Aggregate features per agent
agent_features = df_train.groupby('agent_code').agg({
    'unique_proposal': 'mean',
    'unique_quotations': 'mean',
    'unique_customers': 'mean',
    'new_policy_count': 'mean',
    'number_of_policy_holders': 'mean'
}).reset_index()

# Scale the data
scaler = StandardScaler()
scaled_features = scaler.fit_transform(agent_features.drop('agent_code', axis=1))

# KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=42)
agent_features['cluster'] = kmeans.fit_predict(scaled_features)

# Visualize clusters
sns.pairplot(agent_features, hue='cluster', corner=True)
plt.suptitle("Agent Clustering", y=1.02)
plt.show()
```

### 🧠 Agent Clustering Insights

#### 🧩 Distinct Cluster Profiles
- Agents have been grouped into **three distinct clusters** (`0`, `1`, `2`), each exhibiting **different behavioral patterns** across key performance metrics.
  
- **Cluster 2** (darkest shade) tends to have **higher values** for:
  - `unique_customers`
  - `new_policy_count`
  - `number_of_policy_holders`  
  This suggests that Cluster 2 contains the **high-performing agents**—those excelling in **customer acquisition and policy conversion**.

- **Cluster 0** (lightest shade) generally demonstrates **lower performance** across most features, indicating **lower engagement or effectiveness**.

- **Cluster 1** (intermediate shade) represents a **middle-ground**, with performance metrics typically falling **between Clusters 0 and 2**—likely **average or mid-tier agents**.

#### 🔗 Correlation Observations
- Strong positive relationships are observed between:
  - `unique_customers` and `new_policy_count`
  - `unique_proposals` and `unique_quotations`  
  These correlations suggest that agents who **engage more customers** or **submit more proposals** tend to **convert more policies** and generate more quotations.

#### 📊 Feature Distributions
- The **diagonal plots** in the pairwise visualization show that **most features are right-skewed**, especially:
  - `number_of_policy_holders`  
  This skewness highlights that a **small subset of top agents** likely **drive the bulk of activity**, a common pattern in sales-driven environments.

The plot visualizes agent performance using standardized metrics:

- **X-axis:** Z-score of `New Policy Count`  
- **Y-axis:** Z-score of `Unique Customers`
